```python
from langchain.document_loaders import JSONLoader
import json

# Define the jq schema for extraction
jq_schema = '.[] | {"content": .summary, "metadata": {"doc_name": .doc_name, "page_number": .page_number}}'

# Initialize the loader
loader = JSONLoader(
    file_path="/Users/galbraithelroy/Documents/jacie/EduFocal-Limited-2023-Financial-Statements_summaries.json",
    jq_schema=jq_schema,
    text_content=False
)

docs = loader.load()

docs[0]
```

```python
from langchain_ollama import OllamaLLM, OllamaEmbeddings

model = OllamaLLM(model="llama3.2-vision")
embeddings = OllamaEmbeddings(model="llama3.2-vision")
```

```python
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS
import faiss

index = faiss.IndexFlatL2(len(embeddings.embed_query("hello world")))

vector_store = FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)

vector_store.add_documents(docs)

retriever = vector_store.as_retriever()
```

```python
query = "What are the principle activities of Edufocal Limited?"
retrieved_docs = retriever.get_relevant_documents(query)
retrieved_docs[0]
```

```python
import base64

def encode_image(image_path):
    """Convert an image to base64 string."""
    try:
        with open(image_path, "rb") as img_file:
            return base64.b64encode(img_file.read()).decode('utf-8')
    except Exception as e:
        print(f"Error encoding image {image_path}: {str(e)}")
        return None
```

```python
PAGE_SUMMARY_PROMPT = """
    You are an an experienced financial analyst tasked with summarizing a page of a financial statement.
    The page is a summary of the financial statement for a company. I'll provide you with an image of the page and the document name.
    Your job is to summarize the page in a way such that I know what the page is about.

"""
```

```python
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.output_parsers import StrOutputParser

def generate_qa_pair(image_path, doc_name):
    # convert image to base64
    image_base64 = encode_image(image_path)
    
    # create messages for the model
    system_message = SystemMessage(content=PAGE_SUMMARY_PROMPT)
    prompt = HumanMessage(
        content=[
            {"type": "text", "text": f"Document name: {doc_name}"},
            {"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{image_base64}"}}
        ]
    )
    
    # create chain and invoke
    chain = model | StrOutputParser()
    response = chain.invoke([system_message, prompt])
    return response
```

```python
img = "/Users/galbraithelroy/Documents/jacie/docs/pdf_images/EduFocal-Limited-2023-Financial-Statements_page_3.jpg"
doc_name = "EduFocal-Limited-2023-Financial-Statements.pdf"
```

```python
summary = generate_qa_pair(img, doc_name)
print(summary)
```
